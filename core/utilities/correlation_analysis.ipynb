```python
# Creating correlation plots between employee similarity metrics and usage similarity metrics
# to investigate whether or not correlation exists, helping to justify hypothesis.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
metric_data=pd.read_csv("../../../data/more_metric_data.csv")
```

```python
metric_data.head()
```

### Plotting distance vs. usage for all sampled pairs in dataset.

```python
sns.set(color_codes=True)
sns.regplot(x="distance", y="usage_similarity", data=metric_data)
plt.xlim(1,6)
plt.ylim(0,100)
plt.savefig('../../../data/correlationplots/metrics_plot_no_reports.png')
```
